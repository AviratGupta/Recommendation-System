```python
import pandas as pd
import numpy as np
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
movies.head(1)
```

```python
credits.head(1)
```

```python
movies = movies.merge(credits,on="title")
```

```python
movies = movies[["movie_id","title","overview","genres","keywords","cast","crew"]]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0]
```

```python
movies.iloc[0]["genres"]
```

```python
import ast
def convert(obj):
  arr = []
  for i in ast.literal_eval(obj):
    arr.append(i["name"])
  return arr
```

```python
movies["genres"] = movies["genres"].apply(convert)
```

```python
movies.head(1)
```

```python
movies["keywords"] = movies["keywords"].apply(convert)
```

```python
movies.head()
```

```python
def convert_cast(obj):
  arr = []
  cnt = 0
  for i in ast.literal_eval(obj):
    if cnt!=3:
      arr.append(i["name"])
      cnt+=1
    else:
      break
  return arr
```

```python
movies["cast"] = movies["cast"].apply(convert_cast)
```

```python
movies.head()
```

```python
def convert_crew(obj):
  arr = []
  for i in ast.literal_eval(obj):
    if i["job"] == "Director":
      arr.append(i["name"])
      break
  return arr
movies["crew"] = movies["crew"].apply(convert)
```

```python
movies.head()
```

```python
movies["overview"] = movies["overview"].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#removing spaces between the names so that it become a single entity
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies["tags"] = movies["overview"]+movies["genres"]+movies["cast"]+movies["crew"]+movies["keywords"]
```

```python
new_df = movies[["movie_id","title","tags"]]
```

```python
new_df.head()
```

```python
  new_df["tags"] =  new_df["tags"].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#convert it to lower case
new_df["tags"] = new_df["tags"].apply(lambda x:x.lower())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = "english")
```

```python
vectors = cv.fit_transform(new_df["tags"]).toarray()
```

```python
cv.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
new_df["tags"] = new_df["tags"].apply(stem)
```

```python
#now we calculate distance between 2 vector using cosine distance but not euclidean distance because it is not reliable.
#cosine distance is theta between 2 vectors
#we do it with help of sklearn
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x :x[1])[1:11]
```

```python
def recommend(movie):
  movie_index = new_df[new_df["title"] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x :x[1])[1:11]
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```
